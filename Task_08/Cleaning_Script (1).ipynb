```python
import pandas as pd
```

```python
df = pd.read_csv("superstore_sales.csv", encoding='latin1')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed', dayfirst=False)
df['Ship Date']  = pd.to_datetime(df['Ship Date'],  format='mixed', dayfirst=False)

df[['Order Date', 'Ship Date']].head()
```

```python
df['Month_Year'] = df['Order Date'].dt.to_period('M').astype(str)
df['Month_Year'].head()
```

```python
duplicates_count = df.duplicated().sum()
print(duplicates_count)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
negative_sales = df[df['Sales'] < 0].shape[0]
print(negative_sales)
```

```python
df.describe(include='all')
```

```python
df_clean = df.sort_values(by="Order Date")
df_clean.to_csv("superstore_clean.csv", index=False)

df_clean.head()
```

```python
import zipfile

zip_filename = "superstore_clean.zip"

with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipf.write("superstore_clean.csv")
```
